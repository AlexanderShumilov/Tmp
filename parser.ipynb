```python
def read_seqres(path_to_file):
    
        with open(path_to_file, "r") as file_pdb:
            lines = file_pdb.readlines()
        
        seqres_list = []
        for line in lines:
            if line.startswith("SEQRES"):
                seqres_list.extend(line.split()[4:]) 
        
        return {k: v for k, v in enumerate(seqres_list, 1)}    
  
def read_res(path_to_file):
    
        with open(path_to_file, "r") as file_pdb:
            lines = file_pdb.readlines()
        
        res_list = {}
        pred_res = -100
        flag_comment = False
        
        for line in lines:
            if line.startswith("ATOM") and line.split()[5] != pred_res and '.' not in line.split()[5]:
                #print(line.split())
                if (line.split()[5]).isdigit():
                    res_list[int(line.split()[5])] = line.split()[3]
                    pred_res = int(line.split()[5])
                else: 
                    flag_comment = True
                
        return res_list, flag_comment
'''
def count_dissimilarity(seqres_list, res_list):
    
    count = 0
    mismatch_array = np.zeros(len(seqres_list))
    for i in range(len(res_list)):
        for j in range(i + count, len(seqres_list)):
            if res_list[i] != seqres_list[j]: 
                count += 1
                mismatch_array[j] = 1
            else:
                break
                
    return mismatch_array, count    
'''    
    
def seq_compare(path):

    seqres_list = read_seqres(path)
    res_list, flag_comment = read_res(path)

    
    return seqres_list, res_list, flag_comment

def find_missing_residues(path):
    
    with open(path, "r") as file_pdb:
            lines = file_pdb.readlines()
    
    missing_res = {}
    flag_comment = False
    
    remark = "REMARK" + " " + str(-1)
    for i, line in enumerate(lines):

        if "MISSING RESIDUES" not in line and remark not in line:
            continue
        else:
            remark = "REMARK" + " " + str(line.split()[1])
    
        if  any(aa in line for aa in amino_acids_list) and len(line.split()) == 5:
            if (line.split()[4]).isdigit() or ((line.split()[4][0] == "-") and (line.split()[4][1:].isdigit())):
                missing_res[int(line.split()[4])] = line.split()[2]
            else:
                flag_comment = True
            
    return missing_res, flag_comment
```

```python
files_wt = os.listdir(dir_pdb)
files_pdb = list(filter(lambda x: x.endswith('.ent'), files_wt))
```

```python
df = pd.DataFrame(columns = ["Filename", 
                             "Header info (y/n)", 
                             "N of missing residues", 
                             "Missing residues", 
                             "Seqres matches pdb res_list", 
                             "Comment"])

k=0

for filename in tqdm.tqdm_notebook(files_pdb):
    
    #k +=1
    #if k>5:break
    
    
    flag_header = False
    flag_corresp = True
    data_temp = []
    path = dir_pdb + '/' + filename
    dict_missing_res = {}
    comment = 'Ok'
    flag_comment_header = False
    flag_comment_body = False
    
    with open(path, "r") as handle:
        header_dict = parse_pdb_header(handle)
        handle.close()
        
    if 'has_missing_residues' in header_dict:
        if header_dict['has_missing_residues']:
            
            flag_header = True
            
            dict_missing_res, flag_comment_header = find_missing_residues(path)
            seqres_list, res_list, flag_comment_body = seq_compare(path)

            N_missing = len(dict_missing_res)
            
            if N_missing != abs(len(seqres_list) - len(res_list)) : flag_corresp = False
                        
        else: 

            seqres_list, res_list, flag_comment_body = seq_compare(path)
            N_missing = abs(len(seqres_list) - len(res_list))
            
    #list_seqres.append(seqres_list)
    #list_res.append(res_list)
    #list_miss.append(dict_missing_res)
    
    if flag_comment_header: 
                comment = 'Error in parsing header'
    if flag_comment_body: 
                comment = 'Error in parsing body'
    
    data_temp.append({"Filename": filename, 
                      "Header info (y/n)": flag_header, 
                      "N of missing residues": N_missing,
                      "Missing residues": dict_missing_res,
                      "SEQRES (dict)": seqres_list,
                      "Residues (dict)": res_list,
                      "SEQRES (len)": len(seqres_list),
                      "Residues (len)": len(res_list),
                      "Seqres matches pdb res_list": flag_corresp,
                      "Comment": comment}
                    )
    df_temp = pd.DataFrame(data_temp)
    df = df.append(df_temp, ignore_index = True)
    
```
